```python
#Open the files saved in the dataset directory
import os
import json
file_path = os.path.join('..', 'dataset','Compiled', 'Jan', 'dataset_compiled_jan.json')
with open(file_path) as result_jan:
    result_jan = json.load(result_jan)
file_path = os.path.join('..', 'dataset','Compiled', 'Apr', 'dataset_compiled_apr.json')
with open(file_path) as result_apr:
    result_apr = json.load(result_apr)

file_path = os.path.join('..', 'dataset','Compiled', 'Sep', 'dataset_compiled_sep.json')
with open(file_path) as result_sep:
    result_sep = json.load(result_sep)   
```

```python
# Clean the result from dictionaries with missing "itemList"

result_jan = [v for v in result_jan if 'itemList' in v and v['itemList']]
result_apr = [v for v in result_apr if 'itemList' in v and v['itemList']]
result_sep = [v for v in result_sep if 'itemList' in v and v['itemList']]
```

```python
from pprint import pprint
pprint(result_sep[0])
```

```python
#expanding nestled dictionaries

extended_list_jan = []
extended_list_apr = []
extended_list_sep = []
for i, v in enumerate(result_jan):
    itemlist_jan = v.get('itemList')
    if itemlist_jan:
        extended_list_jan.extend(itemlist_jan)
    else:
        print(f'itemlist missing: {i}')

for i, v in enumerate(result_apr):
    itemlist_apr = v.get('itemList')
    if itemlist_apr:
        extended_list_apr.extend(itemlist_apr)
    else:
        print(f'itemlist missing: {i}')   
        
for i, v in enumerate(result_sep):
    itemlist_sep = v.get('itemList')
    if itemlist_sep:
        extended_list_sep.extend(itemlist_sep)
    else:
        print(f'itemlist missing: {i}')        
```

```python
#Expand dictionary for Author Statistic
author_stats_jan = [v['authorStats'] for v in extended_list_jan]
author_stats_apr = [v['authorStats'] for v in extended_list_apr]
author_stats_sep = [v['authorStats'] for v in extended_list_sep]
```

```python
#Pretty print the author object

from pprint import pprint
pprint(result_sep[0]['itemList'])
```

```python
#Creating dataframes with AuthorsStats for Jan
import json
import pandas as pd
df_jan = pd.DataFrame(extended_list_jan)
df_author_stats_jan = pd.DataFrame(author_stats_jan)
df_jan = df_jan.join(df_author_stats_jan)
df_jan = df_jan.drop('authorStats', axis = 1) #axis=1 columns in the table
```

```python
#select and expand all keys of dictionary 'Author' from the Itemlist 
df_jan['avatarLarger'] = df_jan.author.apply(lambda x: x.get('avatarLarger'))
df_jan['avatarMedium']= df_jan.author.apply(lambda x: x.get('avatarMedium'))
df_jan['avatarThumb']= df_jan.author.apply(lambda x: x.get('avatarThumb'))
df_jan['commentSetting']= df_jan.author.apply (lambda x: x.get('commentSetting'))
df_jan['downloadSetting']= df_jan.author.apply(lambda x: x.get('downloadSetting'))
df_jan['duetSetting']= df_jan.author.apply(lambda x: x.get('duetSetting'))
df_jan['ftc'] = df_jan.author.apply(lambda x: x.get('ftc'))
df_jan['authorId'] = df_jan.author.apply(lambda x: x.get('id'))
df_jan['isADVirtual'] = df_jan.author.apply(lambda x: x.get('isADVirtual'))
df_jan['isEmbedBanned'] = df_jan.author.apply(lambda x: x.get ('isEmbedBanned'))
df_jan['nickname'] =df_jan.author.apply(lambda x: x.get ('nickname'))
df_jan['openFavorite'] = df_jan.author.apply(lambda x: x.get ('openFavorite'))
df_jan['privateAccount'] = df_jan.author.apply(lambda x: x.get ('privateAccount'))
df_jan['relation'] = df_jan.author.apply(lambda x: x.get ('relation'))
df_jan['secUid'] = df_jan.author.apply(lambda x: x.get('secUid'))
df_jan['secret'] = df_jan.author.apply(lambda x: x.get('secret'))
df_jan['signature'] = df_jan.author.apply(lambda x: x.get('signature'))
df_jan['stitchSetting'] = df_jan.author.apply(lambda x: x.get('stitchSetting'))
df_jan['ttSeller'] = df_jan.author.apply(lambda x: x.get('ttSeller'))
df_jan['verified'] = df_jan.author.apply(lambda x: x.get('verified'))
df_jan['authorUniqueId'] = df_jan.author.apply(lambda x: x.get('uniqueId'))

# selecting keys from AuthorStats
df_jan['authorDiggCount'] = df_jan.diggCount
df_jan['authorFollowerCount'] = df_jan.followerCount
df_jan['authorFollowingCount'] = df_jan.followingCount
df_jan['authorHeartCount'] = df_jan.heartCount
df_jan['authorVideoCount'] = df_jan.videoCount
```

```python
#conduct Author dataframe for January 
df_jan_author = df_jan[
['authorId', #ID number of the profile
'authorUniqueId', #The handle of the profile
'avatarLarger',  #A full-size version of the profile photo (link will have a quick expiration date)
'authorDiggCount', #The number of videos liked by the user
'authorFollowerCount',
'authorFollowingCount',
'authorVideoCount', 
'authorHeartCount',
'commentSetting', # who can comment on the user content: 0 = Everyone; 1= Friends; 2= Off
'downloadSetting',# 0: Allow anyone to download videos, 1: Allow friends to download videos, 2: Do not allow anyone to download videos, 3: Allow downloads only in specific regions
'duetSetting', # 0:Allow anyone to duet videos; 1: Allow friends to duet videos; 2: Do not allow anyone to duet videos
'ftc', #False/True, indicates minors under 13 yo. 
'id', #Creator Id
'isADVirtual', #False/True for virtual influenser, not a real person
'nickname', #Text string, The display name of the profile
'openFavorite', #False/True, the user has allowed other users to see their bookmarked videos, sounds, and effects on the web version of TikTok
'privateAccount', #False/True, the user's videos and content are/are'nt accessible to the general public
'relation', #0/1 Family pairing feature
'secUid', # unique string of characters to identify the user or their content
'secret',#False/True, account's content is limited in visibility to a select audience or followers
'signature',
'stitchSetting', #0/1 the creator enabled stitches of his contributions
'ttSeller', #false/true, the creator is selling a product
'verified' #False/True
    ]].copy()

# Get the unique 'authorId' rows from df_jan_cleaned
df_jan_author_unique = df_jan_author.drop_duplicates(subset='authorId')

#Number of unique AuthorID
print('Number of Unique AuthorID in Jan dataset', len(df_jan_author_unique))

#Save to CSV file
df_jan_author_unique.to_csv('Author_df_jan.csv', index=False)
```
